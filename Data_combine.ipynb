```python
import pandas as pd
```

```python
data_hr1 = pd.read_csv('data/employee_survey_data.csv')
```

```python
data_hr1 = data_hr1.drop(["YearsAtCompany"], axis=1)
```

```python
data_hr1
```

```python
data_hr2 = pd.read_csv('data/general_data.csv')
```

```python
data_hr2
```

```python
data_hr4 = pd.merge(data_hr1, data_hr2, on=['EmployeeID'])
```

```python
data_hr4
```

```python
data_hr3 = pd.read_csv('data/manager_survey_data.csv')
```

```python
data_hr3
```

```python
data_hr5 = pd.merge(data_hr4, data_hr3, on=['EmployeeID'])
```

```python
data_hr5
```

```python
data_hr5 = data_hr5.drop(["EmployeeID"], axis=1)
```

```python
data_hr5.columns
```

```python
data_hr5
```

```python
data_hr5 = data_hr5.reindex(sorted(data_hr5.columns), axis=1)
```

```python
data_hr5
```

```python
data_IBM = pd.read_csv('data/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
data_IBM
```

```python
data_IBM = data_IBM.drop(["DailyRate", "HourlyRate", "MonthlyRate", "OverTime", "YearsInCurrentRole", "RelationshipSatisfaction", "EmployeeNumber"], axis=1)

```

```python
data_IBM
```

```python
data_IBM = data_IBM.reindex(sorted(data_IBM.columns), axis=1)
```

```python
data_IBM
```

```python
data_final = pd.concat([data_hr5, data_IBM], ignore_index=True)
```

```python
data_final
```

```python
#data_final.to_csv("Employee_Years.csv",index=False)
```

```python
age_groups = [0, 10, 20, 30, 40, 50, 60, 70, 100]

age_group_names = ['<10', '10-20', '21-30', '31-40', '41-50', '51-60', '61-70', '>70']
```

```python
data_final['Age_group'] = pd.cut(data_final['Age'], bins=age_groups, labels=age_group_names)
```

```python
data_final
```

```python
data_final.to_csv("Employee_Years.csv",index=False)
```
